```python
# Define the board as a list of 9 elements
board = [" ", " ", " ",
         " ", " ", " ",
         " ", " ", " "]

# Define winning positions as a list of tuples
winning_positions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),  # Horizontals
                     (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Verticals
                     (0, 4, 8), (2, 4, 6)]  # Diagonals

# Define player symbols
player = "X"
opponent = "O"

# Define a function to display the board
def display_board():
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} ")

# Define a function to check if the board is full
def is_board_full():
    return " " not in board

# Define a function to check if there is a winner
def check_winner():
    for a, b, c in winning_positions:
        if board[a] == board[b] == board[c] != " ":
            return board[a]
    return None

# Define a function to evaluate the state of the board
def evaluate():
    winner = check_winner()
    if winner == player:
        return 10
    elif winner == opponent:
        return -10
    else:
        return 0

# Implement the Min-Max algorithm with alpha-beta pruning
def min_max(is_max, alpha, beta):
    value = evaluate()
    if value != 0:
        return value
    if is_board_full():
        return 0
    if is_max:
        best_value = float("-inf")
        for i in range(9):
            if board[i] == " ":
                board[i] = player
                value = min_max(False, alpha, beta)
                board[i] = " "
                best_value = max(best_value, value)
                alpha = max(alpha, best_value)
                if beta <= alpha:
                    break
        return best_value
    else:
        best_value = float("inf")
        for i in range(9):
            if board[i] == " ":
                board[i] = opponent
                value = min_max(True, alpha, beta)
                board[i] = " "
                best_value = min(best_value, value)
                beta = min(beta, best_value)
                if beta <= alpha:
                    break
        return best_value

# Modify the evaluate function to consider squares
def evaluate():
    winner = check_winner()
    if winner == player:
        return 10
    elif winner == opponent:
        return -10
    else:
        value = 0
        for a, b, c in winning_positions:
            if board[a] == board[b] == board[c] == player:
                value += 1
            elif board[a] == board[b] == board[c] == opponent:
                value -= 1
        return value

# Modify the best_move function to consider squares
def best_move():
    best_value = float("-inf")
    best_position = -1
    alpha = float("-inf")
    beta = float("inf")
    for i in range(9):
        if board[i] == " ":
            board[i] = player
            value = min_max(False, alpha, beta)
            board[i] = " "
            if value > best_value:
                best_value = value
                best_position = i
            alpha = max(alpha, best_value)
    return best_position

# Define a function to play the game
def play_game():
    # Display the initial board
    display_board()
    # Repeat until the game ends
    while True:
        # Ask the user to choose a valid position
        row = int(input("Choose a row (1, 2, 3): ")) - 1
        col = int(input("Choose a column (1, 2, 3): ")) - 1
        position = row * 3 + col

        while not (0 <= row < 3 and 0 <= col < 3) or board[position] != " ":
            print("Invalid position. Choose another.")
            row = int(input("Choose a row (1, 2, 3): ")) - 1
            col = int(input("Choose a column (1, 2, 3): ")) - 1
            position = row * 3 + col

        # Place the opponent's symbol in the chosen position
        board[position] = opponent
        # Display the updated board
        display_board()
        print("")
        # Check if the opponent has won or if the board is full
        if check_winner() == opponent:
            print("You won. Congratulations!")
            break
        if is_board_full():
            print("It's a draw!")
            break
        # Find the best move for the player
        position = best_move()
        row = position // 3
        col = position % 3
        # Place the player's symbol in the chosen position
        board[position] = player
        # Display the updated board
        display_board()
        # Check if the player has won or if the board is full
        if check_winner() == player:
            print("I've won. Thanks for playing!")
            break
        if is_board_full():
            print("It's a draw!")
            break

# Call the play_game function to start the game
play_game()
```
